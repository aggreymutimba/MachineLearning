## Three-class Generalization of Logistic Regression.

```python
#calculate the sigmoid function
def sigmoid_function(theta_transpose_x):
    return 1 / (1 + np.exp(- theta_transpose_x))
```

```python
x = range(-10, 10, 1)
plt.plot(x , [sigmoid_function(x) for x in x])
plt.title("Figure 1: Sigmoid Function", fontsize=20);
```

```python
for i in set(fruits.fruit):
    color1 = ["y", "r", "b", "g"]
    clf = linear_model.LinearRegression()
    x = fruits[["width", "height"]].values
    y = [1 if y == i else 0 for y in fruits.fruit.values]
    clf.fit(x, y)
    fruits[i] = sigmoid_function(clf.predict(fruits[["width", "height"]].values))
```

```python
fruits["the_max"] = [1 + fruits[[3, 4, 5]][fruits.index == i].values.argmax() for i in fruits.index]

for i in set(fruits.fruit):
    label = {1: "Apple", 2: "Orange", 3 : "Lemon"}
    color0 = ["y", "r", "b", "g"]
    color1 = ["yo", "ro", "bo", "go"]
    color2 = ["yx", "rx", "bx", "gx"]
    plt.plot(fruits[fruits.fruit == i].width, fruits[fruits.fruit == i].height, color1[i], label=label[i]);
    plt.plot(0.05 + fruits[fruits.the_max == i].width, fruits[fruits.the_max == i].height, color2[i],
             label="Classified " + label[i])    
    plt.xlabel("Width (cm)", fontsize=18)
    plt.ylabel("Height (cm)", fontsize=18)
    plt.title("Figure 2: Logistic Regression", fontsize=20)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

## Generative Classication with Gaussian Class-conditional Densities.

```python
fruits = pd.read_csv("fruit.csv")
```

```python
def get_priors(df):
    
    some_dict = {}
    sums = df.groupby("fruit").size()
    
    return sums * 100.0 / np.sum(sums)
    
priors = get_priors(fruits)
```

```python
def likelihood_estimates(f, df):
    
    df = df[df.fruit == f]
        
    stuff = np.vstack((np.array(df.width.values- df.width.mean()), np.array(df.height.values- df.height.mean())))
    
    return np.transpose(np.dot(stuff, stuff.T)/len(df))
```

```python
def conditional_densities(f, value, var, df):
    
    df = df[df.fruit == f]
    
    sigma = likelihood_estimates(f, df)
    
    stuff = np.vstack((np.array(value - df[var].mean()), np.array(value - df[var].mean())))
    
    return (2 * math.pi)**(len(df)/2) * np.linalg.det(sigma)**1/2 * \
            np.exp(-0.5 * np.dot(np.dot(stuff.T, np.linalg.inv(sigma)), stuff))[0][0]
```

```python
max_row = []
for x in fruits[["width", "height"]].values:  
    the_max = -np.inf
    index = 0
    for i in set(fruits.fruit):    
        new_the_max = priors[i] * conditional_densities(i, x[0], "height", fruits) * \
                conditional_densities(i, x[1], "width", fruits)
            
        if new_the_max > the_max:
            the_max = new_the_max
            index = i
            
    max_row.append(index)
    
fruits["max_row"] = max_row
```

```python
for i in set(fruits.fruit):
    label = {1: "Apple", 2: "Orange", 3 : "Lemon"}
    color0 = ["y", "r", "b", "g"]
    color1 = ["yo", "ro", "bo", "go"]
    color2 = ["yx", "rx", "bx", "gx"]
    plt.plot(fruits[fruits.fruit == i].width, fruits[fruits.fruit == i].height, color1[i], label=label[i]);
    plt.plot(0.05 + fruits[fruits.max_row == i].width, fruits[fruits.max_row == i].height, color2[i],
             label="Classified " + label[i])    
    plt.xlabel("Width (cm)", fontsize=18)
    plt.ylabel("Height (cm)", fontsize=18)
    plt.title("Figure 4: Gaussian class-conditional densities", fontsize=20)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
X = fruits[["width", "height"]].values
y = fruits.fruit.values
clf = LDA()
clf.fit(X, y, tol=0.0001)

fruits["LDA"] = clf.predict(X)

for i in set(fruits.fruit):
    label = {1: "Apple", 2: "Orange", 3 : "Lemon"}
    color0 = ["y", "r", "b", "g"]
    color1 = ["yo", "ro", "bo", "go"]
    color2 = ["yx", "rx", "bx", "gx"]
    plt.plot(fruits[fruits.fruit == i].width, fruits[fruits.fruit == i].height, color1[i], label=label[i]);
    plt.plot(0.05 + fruits[fruits.LDA == i].width, fruits[fruits.LDA == i].height, color2[i], 
             label="Classified " + label[i])     
    nx, ny = 200, 200
    x_min, x_max = plt.xlim()
    y_min, y_max = plt.ylim()
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, nx), np.linspace(y_min, y_max, ny))
    Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])
    Z = Z[:, 1].reshape(xx.shape)
    
    plt.contour(xx, yy, Z, [0.5], linewidths=1.75, colors="k")    
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)    
    plt.xlabel("Width (cm)", fontsize=18)
    plt.ylabel("Height (cm)", fontsize=18)
    plt.title("Figure 3: Naive Bayers Classifier", fontsize=20)
```
